```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Python code to read data csv files for null or abstract values that can not be used. The data will be cleaned, condensed, and formated using pandas and numpy functions. The Season_Stats.csv will be the dat file used in this analysis. The 'Seasons_Stats.csv' file contains all season year, player, name, position, team, and point total for each season.

The raw data is displayed below with some 'NaN' values starting with the first rows.

```python
season_stats = pd.read_csv('Seasons_Stats.csv')
season_stats.head()
```

Unneeded statistic column categories is removed from the data frame. Null undefined data (NaN) is also removed from season stats data frame. The season year, player point total, and player age are changed from a float value to an integer value:

```python
seasonStats = season_stats.drop(columns=['GS', 'MP','G','FG','FGA','FG%','2P','2PA','2P%','PER','ORB','ORB%','3PAr','DRB','DRB%','TRB','TRB%','AST%','STL','STL%','BLK','BLK%','TOV','TOV%','TS%','USG%','blanl','blank2','WS','WS/48','OBPM','BPM','VORP','3P','3P%','3PA','DBPM','OWS','DWS','FTr','FT','FTA','FT%','PF','eFG%','AST'])

seasonStats = seasonStats.dropna()
seasonStats['Year'] = seasonStats['Year'].astype(int)
seasonStats['PTS'] = seasonStats['PTS'].astype(int)
seasonStats['Age'] = seasonStats['Age'].astype(int)
```

All players from seasons 1950 - 2017 are put into a dictionary. (The code was commented do to library size. Function will be used to created condensed data results.)

```python
#playerData = pd.DataFrame(seasonStats, columns = ['Year','Player','Pos','Age','Tm','PTS'])
#playerDatalib = playerData.to_dict('index')
```

After removing unneeded data columns, null, void values, the data is sorted by year 1950 to 2017. All players with complete statistics for the categories, player ('Player'), position ('Pos'), age ('Age'), team ('Tm'), and points ('PTS') is displayed sort by year 1950 - 2017.

```python
seasonByYear = seasonStats.sort_values(by = 'Year')
seasonByYear.head()
```

The code below tests to make sure that player statistics are displayed for 1 specific season. The test season is 1963. The goal is to display the top 2 scorers for season year 1950 - 2017.

```python
seasonByYear.loc[seasonByYear['Year'] == 1963]
```

The data below uses the sort values function to organize the top scorers. The top scorers are organized from most to least for all season years 1950 - 2017. This organizes all players from all seasons 1950 -2017 in the data file and displays each player by point total highest to lowest. This data will be represent in a scatter point graph.

```python
allPlayers = seasonByYear.sort_values(by = 'PTS', ascending = False)
allPlayers
```

All NBA scores scatterpoint graph.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=allPlayers,x='Year',y='PTS')
plt.title('NBA All players scores 1950 - 2017') #title
plt.xlabel('Season Year') #x label
plt.ylabel('Points Scored') #y label
plt.show()
```

The code below sorts the player by each season 1950 - 2017 and displays the top 2 scorers for each season 1950 - 2017.

```python
seasonTopScorer = []
seasonTopScorer2 = []
for year in range(1950,2018): 
    currentYear = seasonByYear.loc[seasonByYear['Year'] == year]
    pointSort = currentYear.sort_values(by = 'PTS', ascending=False)
    seasonTopScorer.append(pointSort.iloc[[0]])
    seasonTopScorer2.append(pointSort.iloc[[1]])
    print(pointSort.iloc[[0]])
    print(pointSort.iloc[[1]])
```

A compression is made to the top scorer data from to list all players in 1 column category.

```python
topScorers = pd.concat(seasonTopScorer)
topScorers
```

A compression is made to the 2nd top scorer data from to list all players in 1 column category.

```python
topScorers2 = pd.concat(seasonTopScorer2)
topScorers2
```

NBA top scores for each season 1950 - 2017 graphical plot:

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=topScorers,x='Year',y='PTS')
plt.title('NBA top Scorers 1950 - 2017') #title
plt.xlabel('Season Year') #x label
plt.ylabel('Points Scored') #y label
plt.show()
```

NBA runner up top scores for each season 1950 - 2017 graphical plot:

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=topScorers2,x='Year',y='PTS')
plt.title('NBA Runner up top scorers 1950 - 2017') #title
plt.xlabel('Season Year') #x label
plt.ylabel('Points Scored') #y label
plt.show()
```
